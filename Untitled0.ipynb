```python
import nlp_preprocessing as nfx  # 确保安装：pip install nlp-preprocessing
from tqdm import tqdm

def clean_text(text_list, remove_stopwords=True):
    """
    清理文本：
    1. 转小写
    2. 去除特殊字符
    3. 可选：移除停用词
    4. 计算清理后文本的词数

    :param text_list: List[str] - 输入文本列表
    :param remove_stopwords: bool - 是否移除停用词（默认 True）
    :return: Tuple[List[str], List[int]] - 清理后的文本 & 词数统计
    """
    cleaned_text = []
    text_length = []

    for sent in tqdm(text_list, desc="Cleaning text", disable=len(text_list) < 100):
        sent = sent.lower()
        sent = nfx.remove_special_characters(sent)
        if remove_stopwords:
            sent = nfx.remove_stopwords(sent)
        cleaned_text.append(sent)
        text_length.append(len(sent.split()))

    return cleaned_text, text_length
```

```python
import re
import nltk
from tqdm import tqdm
from nltk.corpus import stopwords

nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

def remove_special_characters(text):
    return re.sub(r"[^a-zA-Z0-9\s]", "", text)

def remove_stopwords(text):
    return " ".join([word for word in text.split() if word.lower() not in stop_words])

def clean_text(text_list, remove_stopwords_flag=True):
    cleaned_text = []
    text_length = []

    for sent in tqdm(text_list, desc="Cleaning text", disable=len(text_list) < 100):
        sent = sent.lower()
        sent = remove_special_characters(sent)  # 替代 nfx.remove_special_characters
        if remove_stopwords_flag:
            sent = remove_stopwords(sent)  # 替代 nfx.remove_stopwords
        cleaned_text.append(sent)
        text_length.append(len(sent.split()))

    return cleaned_text, text_length
```

```python
cleaned_train_text,train_text_length=clean_text(train_data.text)
cleaned_test_text,test_text_length=clean_text(test_data.text)
```
